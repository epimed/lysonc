# Data preparation

Objective: reduce data to a list of selected genes

```python
import pandas as pd
```

```python
data_dir = 'data/'
```

## Import clinical data

```python
import warnings

clin_file = 'EpiMed_experimental_grouping_2023.03.16_TCGA-BRCA.xlsx'

with warnings.catch_warnings(record=True):
    warnings.simplefilter("always")
    expgroup = pd.read_excel(data_dir + clin_file, engine='openpyxl', sheet_name=0, index_col=0)
    parameters = pd.read_excel(data_dir + clin_file, engine='openpyxl', sheet_name=1, index_col=0)
```

```python
expgroup.head()
```

```python
parameters.head()
```

## Import data

Import gene expression data of the dataset TCGA-BRCA. Data format: log2(1+FPKM).

```python
data_file = 'expression_data_tcga_brca_TCGA-BRCA_log_fpkm_1250_samples_42851_genes.csv'
data = pd.read_csv(data_dir + data_file, sep=';', index_col='id_gene')
data.head()
```

## Import a gene list

```python
genelist_file = 'gene_list.csv'
genelist = pd.read_csv(data_dir + genelist_file, sep=';', index_col='entrez')
print(genelist.shape)
genelist.head()
```

## Reduce data to the selected genes

```python
common_id_genes = list(set(data.index).intersection(set(genelist.index)))
reduced_data = data.loc[common_id_genes].sort_values(by='gene_symbol')
print(reduced_data.shape)
reduced_data.head()
```

## Drop samples with no data (all values are NA) 

```python
reduced_data.dropna(axis=1, how='all', inplace=True)
print(reduced_data.shape)
```

## Reduce data to samples with available clinical information 

```python
common_samples = sorted(list(set(expgroup.index).intersection(set(reduced_data.columns))))
clean_data = reduced_data[['gene_symbol', *common_samples]]
print(clean_data.shape)
clean_data.head()
```

## Save clean reduced data

```python
data_output_file = 'expression_data_tcga_brca_TCGA-BRCA_log_fpkm_' + str(reduced_data.shape[1]-1) + '_samples_' + str(reduced_data.shape[0]) + '_genes.csv'
clean_data.to_csv(data_dir + data_output_file, sep=';', index=True)
print('Data saved to', data_dir + data_output_file)
```

## Save merged clinical data

```python
clinical = pd.concat([expgroup, parameters], axis=1)
clinical = clinical.loc[common_samples]
print(clinical.shape)
clinical_output_file = 'clinical_TCGA-BRCA.csv'
clinical.to_csv(data_dir + clinical_output_file, sep=';', index=True)
print('Clinical info saved to', data_dir + clinical_output_file)
```
